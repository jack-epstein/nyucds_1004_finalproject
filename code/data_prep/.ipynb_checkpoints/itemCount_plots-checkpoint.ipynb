# Popular Item Plots
This notebook builds plots to highlight the long tail distribution of song popularity in the training data

```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
from os import listdir
from pathlib import Path
```

```python
#set data path
data_dir = Path('../../data/item_usage/')
```

```python
#read in files into a dataframe
items_df = pd.concat(
    pd.read_csv(c, names=['track_id','unique_plays','total_plays'])
    for c in data_dir.glob('*.csv')
)


#sort the data by unique playcounts
items_df = items_df.sort_values(by='unique_plays', ascending=False)

print("shape:",items_df.shape)
items_df.head()
```

```python
#get total number of unique plays
maxt = items_df.unique_plays.sum()

#add column that gets the cumulative percentage of distinct plays
items_df['pct_plays'] = items_df.unique_plays.cumsum()/maxt
```

```python
#get cutoff points for each decile -- how many items take up x% of the cumulative plays
cutoffs = []

for i in np.linspace(.1,1,10):
    num_items = items_df.loc[items_df.pct_plays <= i].shape[0]
    cutoffs.append(num_items)
    
cutoffs
```

```python
line = np.arange(1,items_df.shape[0]+1)

#plot the long tail
plt.figure(figsize=(8,6))

plt.plot(line, items_df.unique_plays)
plt.fill_between(line, items_df.unique_plays, alpha=0.2)
plt.vlines(x=cutoffs[4], ymin=0, ymax=15000, linestyle='dashed', color='red')
plt.vlines(x=cutoffs[8], ymin=0, ymax=500, linestyle='dashed', color='red')
plt.text(cutoffs[4]+1000,15100,'50% of cumulative listeners')
plt.text(cutoffs[8]+1000,510,'90% of cumulative listeners')

plt.yscale('log')
plt.title('Song Popularity - Long Tail')
plt.xlabel('Rank of Song (most played to least)')
plt.ylabel('Number of Distinct Listeners -- log scale')
plt.xlim(0,line.size)

plt.show()
```

```python
#print quantiles for unique plays
for i in np.flip(np.linspace(.1,1,10)):
    print('decile:',round(i,2),' distinct plays:',int(np.quantile(items_df.unique_plays, i)))
```

```python
#print the percentage of songs that have at least x distinct plays
num_songs = items_df.shape[0]
for i in range(1,15):
    num_beneath = items_df.loc[items_df.unique_plays <= i].shape[0]
    print("min plays:",i," pct of songs:","{:.1%}".format(num_beneath/num_songs))
```

```python
#plot the long tail CDF - didn't use in report
plt.figure(figsize=(8,6))

plt.plot(line, items_df.pct_plays)
plt.fill_between(line, items_df.pct_plays, alpha=0.2)
plt.vlines(x=cutoffs[4], ymin=0, ymax=1.1, linestyle='dashed', color='red')
plt.vlines(x=cutoffs[8], ymin=0, ymax=1, linestyle='dashed', color='red')
plt.text(cutoffs[4]+1000,1.1,'50% of cumulative listeners')
plt.text(cutoffs[8]+1000,1,'90% of cumulative listeners')

plt.title('Song Popularity - Cumulative Share of Listeners')
plt.xlabel('Rank of Song (most played to least)')
plt.ylabel('Cumulative Share of Distinct Listeners')
plt.xlim(0,line.size)
plt.ylim(0,1.2)

plt.show()
```
